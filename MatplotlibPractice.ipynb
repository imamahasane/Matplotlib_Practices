Matplotlib is a plotting library for the Python programming language and its numerical mathematics extension NumPy. It provides an object-oriented API for embedding plots into applications using general-purpose GUI toolkits like Tkinter, wxPython, Qt, or GTK

## Matplotlib is a python library used for data visualization
### You can create bar-plots, scatter-plots, histograms and a lot more with matplotlib

# Line Plot

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x = np.arange(1, 21)
x
```

```python
y = x * 2
y
```

```python
plt.plot(x, y)
plt.show()
```

### Adding Title and Labels

```python
plt.plot(x, y)
plt.title("Line Plot")
plt.xlabel("X-label")
plt.ylabel("Y-label")
plt.show()
```

### Changing Line Aesthetics
1. color
2. linestyle
3. linewidth

```python
plt.plot(x, y, color = 'r', linestyle = ':', linewidth = 2)
plt.title("Line Plot")
plt.xlabel("X-label")
plt.ylabel("Y-label")
plt.show()
```

```python
plt.plot(x, y, color = 'r', linestyle = ':', linewidth = 5)
plt.title("Line Plot")
plt.xlabel("X-label")
plt.ylabel("Y-label")
plt.show()
```

### Adding two lines in the same plot

```python
x = np.arange(1, 11)
x
```

```python
y1 = x * 2
y1
```

```python
y2 = x * 3
y2
```

```python
plt.plot(x, y1, color = 'g', linestyle = '-', linewidth = 3)
plt.plot(x, y2, color = 'r', linestyle = ':', linewidth = 4)
plt.title("Line Plot")
plt.xlabel("X-label")
plt.ylabel("Y-label")
plt.grid(True)
plt.show()
```

### Adding sub-plots

2. subplot(1, 2, 1)

# Bar Plot

```python
student = {"Naima": 91, "Samiul": 74, "Imam": 68}
student
```

```python
name = list(student.keys())
value = list(student.values())
```

```python
name
```

```python
value
```

```python
plt.bar(name, value)
plt.show()
```

### Adding Title and Label

```python
plt.bar(name, value)
plt.title("Distribution of the student markes")
plt.xlabel("Names")
plt.ylabel("Markes")
plt.grid(True)
plt.show()
```

### Horizontal Bar Plot

```python
plt.barh(name, value)
plt.title("Distribution of the student markes")
plt.xlabel("Names")
plt.ylabel("Markes")
plt.grid(True)
plt.show()
```

```python
plt.barh(name, value, color = 'b')
plt.title("Distribution of the student markes")
plt.xlabel("Names")
plt.ylabel("Markes")
plt.grid(True)
plt.show()
```

# Scatter Plot

### Creating a basic scatter plot

```python
a = [10, 20, 30, 40, 50, 60, 70, 80, 90]
b = [8, 1, 7, 2, 0, 3, 7, 3, 2]
```

```python
plt.scatter(a, b, color = 'r')
plt.show()
```

### Changing Make Aesthetics

```python
plt.scatter(a, b, marker = "*", color = 'b', s = 200)
plt.show()
```

### Adding Tow Markers in The same plot

```python
x = [10, 20, 30, 40, 50, 60, 70, 80, 90]
m1 = [8, 1, 7, 2, 0, 3, 7, 3, 2]
m2 = [9, 1, 2, 8, 3, 7, 4, 5, 6]
```

```python
plt.scatter(x, m1, marker = "*", color = 'r', s = 200)
plt.scatter(x, m2, marker = ".", color = 'b', s = 200)

plt.show()
```

 ### Adding Sub Plot 

```python
plt.subplot(1, 2, 1)
plt.scatter(x, m1, marker = "*", color = 'r', s = 200)

plt.title("Line Plot")
plt.xlabel("X-label")
plt.ylabel("Y-label")
plt.grid(True)

plt.subplot(1, 2, 2)
plt.scatter(x, m2, marker = ".", color = 'b', s = 200)

plt.title("Line Plot")
plt.xlabel("X-label")
plt.ylabel("Y-label")
plt.grid(True)

plt.show()
```

# Pie Chart

### Creating Data

```python
f = ['Apple', 'Orange', 'Mango', 'Guava']
q = [90, 55, 76, 121]
```

### Making Plot

```python
plt.pie(q, labels = f)
plt.show()
```

### Changing Aesthetics

```python
plt.pie(q, labels = f, autopct = '%0.1f%%', colors = ['yellow', 'grey', 'blue', 'black'])
plt.show()
```
